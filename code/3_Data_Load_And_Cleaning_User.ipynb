```python
#importing important libraries
import pandas as pd
import numpy as np
import json
```

## Converting User Json file to Dataframe

```python
#since pd.readjson will cause memeory error, we read the file line by line
path = '../data/input_json_files/yelp_academic_dataset_user.json'
user = []
with open(path, encoding = 'utf-8') as fin:
    for line in fin:
        line_contents = json.loads(line)
        user.append(line_contents)
```

```python
user = pd.DataFrame(user)
user.shape
```

```python
user.head()
```

```python
#Selecting only user id and name columns
user = user.iloc[:,:2]
```

```python
user.to_csv('../data/output_csv/user_clean.csv',index = False)
```
